<strong>Implementação das Funções</strong>

OBS: As funções abaixa foram implementadas utilizando a estrutura de dicionários do Python

```python
def cria_lista_vazia():
    global lista
    lista = {}


def inserir_elemento_na_lista(no):
    tam_lista = lista.__len__()
    if tam_lista == 0:
        lista[tam_lista + 1] = no
    elif tam_lista == 1:
        if lista[tam_lista]['numero'] < no['numero']:
            lista[tam_lista + 1] = no
            lista[tam_lista]['proximo'] = no
        else:
            lista[tam_lista + 1] = lista[tam_lista]
            lista[tam_lista] = no
            lista[tam_lista]['proximo'] = lista[tam_lista + 1]
    else:
        elem = lista[1]
        anterior = 0
        while elem != {} and elem['numero'] < no['numero']:
            anterior = buscar_chave_elemento(elem['numero'])
            elem = elem['proximo']
        no['proximo'] = lista[anterior]['proximo']
        lista[anterior]['proximo'] = no


def ordena_lista():
    lista_atualizada = {}
    cont = 0
    elem = lista[1]
    while elem != {}:
        lista_atualizada[cont] = elem
        elem = elem['proximo']
        cont = cont + 1
    return lista_atualizada


def cria_no(numero):
    no = {'numero': numero, 'proximo': {}}
    return no


def verifica_lista_vazia():
    if lista.__len__() == 0:
        return 1
    return 0


def buscar_elemento_lista(no):
    for chave in lista:
        if lista[chave]['numero'] == no:
            return lista[chave]


def buscar_chave_elemento(no):
    for chave in lista:
        if lista[chave]['numero'] == no:
            return chave


def retorna_cabeca_lista():
    return lista[0]


def liberar_lista():
    lista.clear()
    print("Lista liberada com sucesso!")


def buscar_no_pai(no):
    for chave in lista:
        if lista[chave]['numero'] == no['numero'] and chave != 0:
            return chave-1
    return 0


def remover_elemento_lista(no):
    no_removido = buscar_elemento_lista(no)
    if no_removido != {}:
        posicao_no_pai = buscar_no_pai(no_removido)
        if posicao_no_pai != 0:
            lista[posicao_no_pai]['proximo'] = no_removido['proximo']
            lista.pop(posicao_no_pai+1)
            print('Nó removido com sucesso!')
        elif posicao_no_pai == buscar_chave_elemento(no):
            lista.pop(posicao_no_pai)
            print('Nó removido com sucesso!')
        else:
            lista[posicao_no_pai]['proximo'] = no_removido['proximo']
            lista.pop(posicao_no_pai + 1)
            print('Nó removido com sucesso!')
    else:
        print('O nó informado não foi encontrado!')


def remover_elemento_lista_recursivamente(no_removido,no_proximo):
    if no_removido['numero'] == no_proximo['numero']:
        posicao_no_removido = buscar_chave_elemento(no_removido['numero'])
        if posicao_no_removido-1 != 0:
            lista[posicao_no_removido-1]['proximo'] = no_removido['proximo']
            lista.pop(posicao_no_removido)
            print("Nó removido com sucesso!")
        else:
            lista.pop(posicao_no_removido)
            print("Nó removido com sucesso!")
        return 0
    else:
        if no_proximo['proximo'] == {}:
            return 0
        remover_elemento_lista_recursivamente(no_removido, no_proximo['proximo'])

def imprime_no(no):
    print(f"Valor: {no['numero']} | Próximo: {no['proximo']}")


def imprime_lista():
    for chave in lista:
        print(f"Posição na lista: {chave} | Valor: {lista[chave]['numero']} | Próximo: {lista[chave]['proximo']} ")


def imprime_lista_ordenada(lista):
    for chave in lista:
        print(f"Posição na lista: {chave} | Valor: {lista[chave]['numero']} | Próximo: {lista[chave]['proximo']} ")


def imprime_lista_recursivamente(no):
    print(f"Posição na lista: {buscar_chave_elemento(no['numero'])} | Valor: {no['numero']} | Próximo: {no['proximo']} ")
    if no['proximo'] == {}:
        return 0
    imprime_lista_recursivamente(no['proximo'])


def imprime_lista_reversa():
    tamanho_lista = lista.__len__()
    while tamanho_lista != 0:
        if lista[tamanho_lista-1]:
            print(f"Valor: {lista[tamanho_lista-1]['numero']} | Próximo: {lista[tamanho_lista-1]['proximo']} ")
            tamanho_lista = tamanho_lista - 1


def verifica_listas_iguais(lista_01, lista_02):

    if lista_01.__len__() == lista_02.__len__():
        no_iguais = 0
        for chave in lista:
            if lista_01[chave]['numero'] == lista_02[chave]['numero'] and lista_01[chave]['proximo'] == lista_02[chave]['proximo']:
                no_iguais = no_iguais + 1
        if no_iguais == lista.__len__():
            print("A duas listas informadas são iguais!")
        else:
            print("A duas listas informadas não são iguais!")
    else:
        print("A duas listas informadas não são iguais!")
```

<strong>Execução das Funções</strong>

```python
# 1
cria_lista_vazia()
# 2
no = cria_no(10)
inserir_elemento_na_lista(no)
no = cria_no(30)
inserir_elemento_na_lista(no)
no = cria_no(20)
inserir_elemento_na_lista(no)
no = cria_no(40)
inserir_elemento_na_lista(no)
no = cria_no(35)
inserir_elemento_na_lista(no)
lista = ordena_lista()
# 3
imprime_lista()
# 4
imprime_lista_recursivamente(retorna_cabeca_lista())
# 5
imprime_lista_reversa()
# 6
status_lista = verifica_lista_vazia()
print(f"Status da Lista: {status_lista}")
# 7
no_busca = buscar_elemento_lista(10)
imprime_no(no_busca)
# 8
remover_elemento_lista(10)
lista = ordena_lista()
imprime_lista()
# 9
no_removido = buscar_elemento_lista(40)
remover_elemento_lista_recursivamente(no_removido, retorna_cabeca_lista())
imprime_lista()
# 10
lista_01 = lista.copy()
lista_02 = lista.copy()
verifica_listas_iguais(lista_01, lista_02)
# 11
liberar_lista()
```
